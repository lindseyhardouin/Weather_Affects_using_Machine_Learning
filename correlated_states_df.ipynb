```python
import pandas as pd
import datetime
```

```python
#import the price data for the specified range, Jan 1,2021 - Feb 28, 2021

price_df = pd.read_csv('blmn_close_prices.csv')

price_df['date'] = price_df['time']

del price_df['time']

price_df.tail()
```

```python
weather_df = pd.read_csv('data/cleaned_data/df_weather_data_updated.csv')
```

```python
corr_states_sliced_df = weather_df.drop(columns=[
    'request.type', 
    'request.language', 
    'request.unit',
    'location.lat',
    'location.lon',
    'location.timezone_id',
    'location.utc_offset',
    ])
corr_states_sliced_df = corr_states_sliced_df.drop(correlated_states_df.columns[[0,1]],axis=1)
corr_states_sliced_df.rename(columns={
    'location.region':'state', 
    'request.query': 'zipcode', 
    'location.name': 'city',
    'location.country':'country'
    }, inplace=True)

date = corr_states_sliced_df.pop('date')
corr_states_sliced_df.insert(0, 'date', date)

state = corr_states_sliced_df.pop('state')
corr_states_sliced_df.insert(1, 'state', state)

corr_states_sliced_df['date']= pd.to_datetime(corr_states_sliced_df['date'], utc=True)#.dt.date

price_df['date'] = pd.to_datetime(price_df['date'], utc=True)#.dt.date

#corr_states_sliced_df = corr_states_sliced_df['state'].isin(correlated_states_list)

corr_states_sliced_df
```

```python
grouped_df = corr_states_sliced_df.groupby(["state", 'date'], as_index=False).mean()

grouped_df.sort_values('date')
```

```python
Arizona = grouped_df[grouped_df['state']=='Arizona']

Florida = grouped_df[grouped_df['state']=='Florida']

Georgia = grouped_df[grouped_df['state']=='Georgia']

Nevada = grouped_df[grouped_df['state']=='Nevada']

South_Carolina = grouped_df[grouped_df['state']=='South Carolina']
```

```python
 Arizona_df = pd.merge(Arizona, price_df, how='outer', on='date').dropna()

 Arizona_df
```
