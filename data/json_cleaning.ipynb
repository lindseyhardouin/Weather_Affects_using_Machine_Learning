```python
import os
import pandas as pd
import json
import csv
from dotenv import load_dotenv
```

```python
#Opening json files with api call responses

f1 = open('api_responses/weather13.json')
f2 = open('api_responses/weather18.json')
f3 = open('api_responses/weather19.json')
f4 = open('api_responses/weather20.json')
f5 = open('api_responses/weather10.json')
```

```python
#Loading json files and saving as responses

responses1 = json.load(f1)
responses2 = json.load(f2)
responses3 = json.load(f3)
responses4 = json.load(f4)
responses5 = json.load(f5)
```

```python
# Calculate the numbers list for loops below based on how many responses in each json file
numbers1 = list(range(0, 20200))
numbers2 = list(range(0,20200))
numbers3 = list(range(0, 14140))
numbers4 = list(range(0, 2020))
numbers5 = list(range(0,3838))


display(len(responses1))
display(len(responses2))
display(len(responses3))
display(len(responses4))
display(len(responses5))


errors = []
responses_clean = []
```

```python
#Run for loops to remove errored out API calls

for number in numbers1:
    x = number
    x1 = len(responses1[x])
    if x1 == 4:
        responses_clean.append(responses1[x])
for number in numbers2:
    x = number
    x1 = len(responses2[x])
    if x1 == 4:
        responses_clean.append(responses2[x])
for number in numbers3:
    x = number
    x1 = len(responses3[x])
    if x1 == 4:
        responses_clean.append(responses3[x])
for number in numbers4:
    x = number
    x1 = len(responses4[x])
    if x1 == 4:
        responses_clean.append(responses4[x])
for number in numbers5:
    x = number
    x1 = len(responses5[x])
    if x1 == 4:
        responses_clean.append(responses5[x])
```

```python
#Begin cleaning data by normalizing top level and creating index to work with
cleaned_data = pd.json_normalize(responses_clean)
cleaned_data
```

```python
#Cleaning hourly data which is average for the day based on input interval in the api call request made

daily_data = []

for response in responses_clean:
    normalize = pd.json_normalize(response['historical'][list(response['historical'].keys())[0]]['hourly'])
    daily_data.append(normalize)
    
daily_data = pd.concat(daily_data)
daily_data['index'] = cleaned_data.index
daily_data = daily_data.set_index('index')
```

```python
daily_data
```

```python
# Cleaning second level of data which has historical information needed but nests it into seperate columns.  This removes the nesting that would happen if just json_normalizing
historical_data = []

for response in responses_clean:
    normalize = pd.json_normalize(response['historical'][list(response['historical'].keys())[0]])
    historical_data.append(normalize)
```

```python
# Connecting data together and providing index
historical_data  = pd.concat(historical_data)
historical_data['index'] = cleaned_data.index
historical_data = historical_data.set_index('index')
```

```python
historical_data
```

```python
#selecting columns wanted from cleaned_data dataframe
cleaned_data_col = [
'request.type',
'request.query',
'request.language',
'request.unit',
'location.name',
'location.country',
'location.region',
'location.lat',
'location.lon',
'location.timezone_id',
'location.utc_offset',
]
```

```python
#Creating df_weather_data
df_weather_data = []
```

```python
#Connecting all 3 tables together 

df_weather_data = pd.concat([cleaned_data[cleaned_data_col], daily_data, historical_data], axis=1)
```

```python
# Last dropping of columns to get data down to desirable data
df_weather_data = df_weather_data.drop(columns=[
 'time',
 'wind_speed',
 'wind_degree',
 'wind_dir',
 'weather_code', 
 'weather_icons',
 'weather_descriptions',
 'chanceofrain',
 'chanceofremdry',
 'chanceofwindy',
 'chanceofovercast',
 'chanceofsunshine',
 'chanceoffrost',
 'chanceofhightemp',
 'chanceoffog',
 'chanceofsnow',
 'chanceofthunder',
 'uv_index',
 'date_epoch',
 'uv_index',
 'astro.sunrise',
 'astro.sunset',
 'astro.moonrise',
 'astro.moonset',
 'astro.moon_phase',
 'astro.moon_illumination',
 'hourly'])
```
